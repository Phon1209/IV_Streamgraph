```python
import numpy as np
import pandas as pd

# Step 1: Read the CSV file and parse it into a pandas DataFrame
file_path = 'Phon_spending.csv'
df = pd.read_csv(file_path)
df['Month'] = pd.to_datetime(df['Month'])
df.set_index(["Month"], inplace=True)

# Step 2: Replace NaN values with random numbers between 100 and 300
random_values = np.random.randint(100, 301, size=df.shape)
df = df.fillna(pd.DataFrame(random_values, index=df.index, columns=df.columns))
```

```python
df_daily = df.resample("D").asfreq()
df_daily = df_daily.fillna(method='ffill')
print(df_daily)
```

```python
cols = df.columns.tolist()
for month_start, month_end in zip(df.index, df.index[1:]):
    for col in cols:
        # Calculate total spending for the current month
        monthly_total = float(df.loc[month_start, col])
        # print(col, month_start, monthly_total)

        # Calculate the number of days in the current month
        num_days = (month_end - month_start).days
        
        # Generate random spending for each day of the month
        random_spending = np.random.dirichlet(np.ones(num_days))
        random_spending = random_spending * monthly_total
        print(random_spending)
        
        # Assign the random spending to the corresponding days
        df_daily.loc[month_start:month_end - pd.Timedelta(days=1), col] = random_spending
```

```python
window_size = 7  # You can adjust this according to your preference
for col in cols:
  df_daily[col] = df_daily[col].rolling(window=window_size, min_periods=1).mean()


df_daily.to_csv("test.csv")
```
